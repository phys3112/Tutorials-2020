## Part one - ODE int!
In this section we will learn how to use ODE int, from the scipy integration package.


```python
import scipy.integrate
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Start out by defining the function we want to integrate.
# We will start slow with an easy function
# Our function HAS TO RECIEVE both y and t, even if you only use one of them!
def fun(y, t):
    # The function dy/dt = 2t (complex!!)
    return 2*t
```

```python
# Begin preparation for integration.

# 1. Generate a 'timebase':
# (Technically it's only a time-base if the x-axis is time, which is the case for most functions we will see.)
timestep = 0.1 #Units? Depends what units your constants are in!
duration = 1000 
timebase = np.arange(duration//timestep)*timestep # The double slash is 'divide but give me an integer only' 
print(timebase)

# 2. Decide on initial conditions
y0 = 5 # Easy starting condition

# 3. Integrate - we are expecting a parabola!
y = scipy.integrate.odeint(fun, y0, timebase)
plt.plot(timebase, y)
```

## Part two - Splitting Everything into 1D ODEs
As we have learned from Michael's lesson, the method of solution for higher order and complex ODEs is simply to set them up as many, many coupled first-order ODEs. This is what we will practice now!

Step 1 is to re-write your equations algebraically in state-variable (or normal) form. Let's assume that we're using a harmonic oscillator, which has two coupled ODEs;
* dx/dt = v
* dv/dt = -kx

Where k is a constant that I can't be bothered writing in LaTeX

```python
def osc(y, t):
    # The tuple 'y' is structured so that y[0] = x, y[1] = v.
    x, v = y # Extract the x and v values from our tuple
    dxdt = v
    dvdt = -5*x
    # For now I have Hard-coded the value for k.
    return dxdt, dvdt
```

```python
# Begin preparation for integration.

# 1. Generate a 'timebase':
# (Technically it's only a time-base if the x-axis is time, which is the case for most functions we will see.)
timestep = 0.1 #Units? Depends what units your constants are in!
duration = 100 
timebase = np.arange(duration//timestep)*timestep # The double slash is 'divide but give me an integer only' 
print(timebase[0:5])

# 2. Decide on initial conditions
x0 = 0.1 
v0 = 0
y0 = x0, v0 #construct our initial conditions tuple

# 3. Integrate - we are expecting a sinusoid!
y = scipy.integrate.odeint(osc, y0, timebase)
x = y[:,0] # Extract our x values from y (x is the zeroth column of y)
plt.plot(timebase, y[:,0])

# Of course, all units here have been arbitrary!
```
